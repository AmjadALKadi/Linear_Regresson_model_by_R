
#  ** simple Linear Regression  lm**

#### H0: no significant differences between Xs and Y
#### H1: there are significant differences between Xs and Y

### H0: B = 0
### H1: B # 0
نقوم بعمل الموديل الخاص بالانحدار الخطي البسيط  
مع اخذ بعين الاعتبار الفرصية الصفرية 
الفرضية الصفرية انه لا يوجد اي اختلاف بين المتغير المدروس والمتغير المؤثر عليه وفق المعادلة 
### Y = a + b * X
Y =  المتغير المدروس
X =  المتغير المؤثر




```r
model <- lm(LungCap ~ Smoke ,data =lung_data)
```

```r
# put this model in summary to view and study it 

model %>% summary()
```

#### coefficients
#### estimat =slop = beta conficient
#### that mean if your are smoking you have  lung capcity more the non-smokine people about 0.87
#### make modle by summer to smoke or non-smoke people


```r
summary(lung_data$LungCap[lung_data$Smoke== 'yes']) # for smoking
```

```r
summary(lung_data$LungCap[lung_data$Smoke== 'no'])   # for non-smoking
```

#### mean(smoke) > mean(non-smoke)
### slop = 0.87

من دراسة تفصيل الموديل نجد ان معادلة الانحدار الخطي 
## lungCap = (Intercept) + slop * Smoke

 ## **MULTI Linear Regression **

#### Multiple: one DV/multiple IVs 
[LungCap = a + B * Smoking + Age + Sex ...]
#### Y = a + ßX + ßX2 +  ßX3 + ...

##### DV = Outcome = Response = Y

##### IV = Exposure = Predictor = Explantory = X

```r
model2 <- lm(LungCap ~ Smoke + Age  + Gender +Height,data= lung_data)
```

```r
model2 %>% summary()
```

##### smoking have lung capacity less then non-smoking about 0.61
##### the older age has more lung capacity about 0.16 then child 
##### the male has lung capacity more about 0.38 from female
##### if you are heighten the lung capacity be increase about 0.26
##### all p_value < 0.05 so h0 reject there are statically significant 
##### R^2 = 85%
##### multi - model its explain  the varty of lung capacity better then model_1

##### lung capacity regression model \

### lung capacity = -10.94 +(-0.61 *smoke)+(0.16*Age)+(-0.38*Gender)+(0.26*Height)
